```python
# Sample leads
leads = [
    {"name": "Rahul", "age": 35, "income": 800000, "family_members": 3, "engagement_score": 70},
    {"name": "Priya", "age": 28, "income": 600000, "family_members": 2, "engagement_score": 40},
    {"name": "Priyanka", "age": 32, "income": 900000, "family_members": 2, "engagement_score": 90},
    {"name": "Shashank", "age": 45, "income": 500000, "family_members": 4, "engagement_score": 20},
    {"name": "Priytam", "age": 27, "income": 600000, "family_members": 3, "engagement_score": 30},
    {"name": "Puneet", "age": 38, "income": 800000, "family_members": 2, "engagement_score": 40},
    {"name": "Raina", "age": 48, "income": 700000, "family_members": 5, "engagement_score": 60},
    {"name": "Arushkaa", "age": 30, "income": 300000, "family_members": 3, "engagement_score": 20},
    {"name": "Tanya", "age": 25, "income": 300000, "family_members": 2, "engagement_score": 20},

]

def calculate_score(lead):
    score = 0
    
    # Age scoring
    if 18 <= lead["age"] <= 50:
        score += 5
    
    # Income scoring
    if lead["income"] > 600000:
        score += 6

    # Family size scoring
    if lead["family_members"] >= 3:
        score += 4

    # Engagement scoring
    if lead["engagement_score"] > 50:
        score += 10

    return score

def assign_priority(score):
    if score >=20:
        return"HIGH"
    elif score>=10:
        return"MEDIUM"
    else:
        return"LOW"

# Score and sort leads
for lead in leads:
    lead["score"] = calculate_score(lead)
    lead["priority"]= assign_priority(lead["score"])

sorted_leads = sorted(leads, key=lambda x: x["score"], reverse=True)

# Output
print("Lead Prioritization Results:")
print("-" * 40)
for lead in sorted_leads:
    print(f"Name: {lead['name']}")
    print(f"Score: {lead['score']} / 25")
    print(f"Priority: {lead['priority']}")
    print("-" * 40)
```
